```python
import torch
from ultralytics.nn.tasks import DetectionModel
import json
with torch.serialization.safe_globals([DetectionModel]):
    ckpt = torch.load(path, map_location="cpu", weights_only=False)

model = ckpt['model']  # Đây là DetectionModel
names = model.names    # trực tiếp từ object
print(names)
with open("classes_v1.json", "w", encoding="utf-8") as f:
    json.dump(names, f, indent=2, ensure_ascii=False)

print("Saved classes.json successfully!")
```

```python
import math
from ultralytics import YOLO

# Load model
model = YOLO("C:\\Users\\ADMIN\\Desktop\\CloneCV\\yolo-object-detection-onnxruntime-web\\trafficSignModel.pt")

# Kích thước thật của biển báo (m) – bạn tự điền theo dataset
SIGN_REAL_HEIGHT = {
    "cam-dung-va-do-xe": 0.6,
    "speed_limit": 0.7,
    "cam": 0.6,
}

FOCAL_LENGTH_PX = 820  # ước lượng hoặc calibrate

def estimate_distance(real_h, pixel_h, focal=FOCAL_LENGTH_PX):
    return (real_h * focal) / pixel_h


results = model("C:\\Users\\ADMIN\\Desktop\\CloneCV\\yolo-object-detection-onnxruntime-web\\anh-bien-cam-dung-xe-01120904.jpg")

for r in results:
    for box in r.boxes:
        cls = int(box.cls[0])
        name = model.names[cls]
        x1, y1, x2, y2 = box.xyxy[0]
        h_pix = y2 - y1

        if name in SIGN_REAL_HEIGHT:
            real_h = SIGN_REAL_HEIGHT[name]
            dist = estimate_distance(real_h, h_pix)
            print(f"{name}: {dist:.2f} m")
```
